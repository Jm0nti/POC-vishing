# Extracción de Features (Vishing Project)

Este notebook contiene un snippet simple para cargar el dataset original aumentado, filtrar las columnas ignoradas por el pipeline de modelado y presentar en una lista las variables predictoras (features) utilizadas.

```python
# 1. Setup and Imports
import pandas as pd

# 2. Load Original Dataset
# Utilizamos la misma ruta S3 definida en el notebook original
s3_parquet_path = 's3://poc-fraude-vishing/proyecto/data/augmented_data/dataset_1M_vishing_.parquet'
df = pd.read_parquet(s3_parquet_path)

# 3. Filter Columns and Extract Features
# Columnas que fueron marcadas para eliminar
cols_to_drop = [
    'session_id', 'customer_id', 'session_timestamp',
    'device_type', 'os_type', 'app_version',
    'biocatch_risk_score', 'biocatch_genuine_score', 'biocatch_ato_indicator', 
    'biocatch_social_eng_indicator', 'biocatch_bot_indicator', 'days_to_claim', 'claim_category',
    'screens_visited', 'unusual_screen_visits', "is_synthetic", "interactions_per_s"
]

# Añadimos además la variable objetivo 'is_vishing' para que no quede en la lista de features
cols_to_drop.append('is_vishing')

# Extraemos la lista final de features retenidas filtrando el dataframe original
features_list = [col for col in df.columns if col not in cols_to_drop]

# Imprimimos la lista
print(f"Total de features: {len(features_list)}\n")
print(features_list)
```
